```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy 
```

```python
def dist_estimation(univariate_data,
                    dist_name):
    size = len(univariate_data)
    x = np.linspace(-4, 4, size)
    print x
    y = univariate_data.as_matrix()
    print y

    dist = getattr(scipy.stats, dist_name)
    param = dist.fit(y.T[0])
    pdf_fitted = dist.pdf(x,
                          *param[:-2],
                          loc=param[-2],
                          scale=param[-1])
   

    return x, pdf_fitted

```

```python
mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
s = pd.DataFrame(s, columns=['Gaussian Noise'])
```

```python
plt.figure()
s.plot(y='Gaussian Noise')
plt.show()
```

```python
plt.plot()
sns.distplot(a=s, kde=False, bins=None, norm_hist=True)
```

```python
x, pd = dist_estimation(s, 'norm')
len(pd)
plt.plot(x, pd)
```
